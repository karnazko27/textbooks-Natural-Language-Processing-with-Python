__19. What is the difference between the following two lines? Which one will give a larger value? Will this be the case for other texts?__

`sorted(set(w.lower() for w in text1))`

`sorted(w.lower() for w in set(text1))`

The first expression lowercases the words first then finds distinct words and symbols. The second expression first finds distinct words and symbols then lowercases them. In the case of the first expression words like 'The' and 'the' will not be repeated since they will be lowercased first. However, with the second expression this error is more likely to occur since the `set()` function will see 'The' and 'the' as two distinct types in the text. Therefore the second expression will be larger.

```python
sorted(set(w.lower() for w in text1))
```

```python
sorted(w.lower() for w in set(text1))
```

```python
# first expression
len(sorted(set(w.lower() for w in text1)))
```

```python
# second expression
len(sorted(w.lower() for w in set(text1)))
```
